SENTIMENT CLASSIFIER

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report
```

```python
# We don't have a positive/negative sentiment column in our dataset. However, this output can be deducted from Rating and Recommended IND.
reviews[['Rating', 'Recommended IND']].value_counts()
# It seems that, in great majority, if an item is ranked 4 or plus, the sentiment is positive because the item is recommended.
# For 1 and 2 rankings, we sense a negative sentiment.
# It is less obvious with rating 3, which works as a neutral rating (half positive, half negative).
```

```python
reviews['Sentiment'] = reviews['Rating'].apply(
    lambda x: 'Positive' if x >= 4 else ('Neutral' if x == 3 else 'Negative')
)
```

```python
reviews_wna = reviews.dropna(subset = 'Review Text').reset_index()
```

```python
X = reviews_wna['Review Text']
y = reviews_wna['Sentiment']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 30)
```

```python
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

Logisitc Regression

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_lr))
```

```python
pd.crosstab(y_test, y_pred_lr, rownames = ['Classes relles'], colnames = ['Classes prdites'])
```

```python
clf = GradientBoostingClassifier(n_estimators = 100, learning_rate = 1, max_depth = 1, random_state = 0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
pd.crosstab(y_test, y_pred, rownames = ['Classes relles'], colnames = ['Classes prdites'])
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
def lemmatization(liste_mots) :
    racines = []
    for mot in liste_mots :
        rac = wordnet_lemmatizer.lemmatize(mot, pos = 'n')
        if rac not in racines :
            racines.append(rac)
    return racines
```

```python
# If we don't include 'neutral' values that are difficult to identify for the model

reviews['Sentiment'] = reviews['Rating'].apply(
    lambda x: 'Positive' if x >= 4 else ('Negative' if x <= 2 else np.nan)
)
reviews_wna = reviews.dropna(subset = ['Review Text', 'Sentiment']).reset_index()

X = reviews_wna['Review Text']
X_lem = X.str.split()
for i in range(len(X_lem)) :
    X_lem[i] = lemmatization(X_lem[i])
    X_lem[i] = ' '.join(X_lem[i])

# Tokenization and Stopwords
X_lem = X_lem.str.lower()

for i in range(len(X_lem)) :
    X_lem[i] = word_tokenize(X_lem[i], language = 'english')

X_tsw = X_lem.apply(stop_words_filtering)

for i in range(len(X_tsw)) :
    X_tsw[i] = ' '.join(X_tsw[i])
    
for i in range(len(X_tsw)) :
    X_tsw[i] = re.sub(r"[\.\+\!\-\,\'s\"]",' ',X_tsw[i])

X = X_tsw
y = reviews_wna['Sentiment']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 30)

vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)

lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

print(classification_report(y_test, y_pred_lr))
pd.crosstab(y_test, y_pred_lr, rownames = ['Classes relles'], colnames = ['Classes prdites'])
```

```python
clf = GradientBoostingClassifier(n_estimators = 100, learning_rate = 1, max_depth = 1, random_state = 0)
clf.fit(X_train, y_train)
y_pred_GB = clf.predict(X_test)

print(classification_report(y_test, y_pred_GB))
pd.crosstab(y_test, y_pred_GB, rownames = ['Classes relles'], colnames = ['Classes prdites'])
```
